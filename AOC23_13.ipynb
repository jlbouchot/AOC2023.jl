```julia
function find_potential_hline(input_pattern)
    return findall([all(x -> x == 0, input_pattern[i+1] - input_pattern[i]) for i in range(1, length(input_pattern)-1)])
end
```

```julia
function find_potential_vline(input_pattern)
    return findall([all([input_pattern[i][j+1] == input_pattern[i][j] for i in range(1, length(input_pattern))]) for j in range(1, length(input_pattern[1])-1)])
end
```

```julia
function validate_vertical_symmetry(input_map, l)
    r = l+2
    l = l-1
    is_valid = true
    while (is_valid) && (l > 0) && (r < length(input_map[1])+1)
        is_valid = all([input_map[j][l] == input_map[j][r] for j in range(1, length(input_map))])
        l -= 1
        r += 1
    end
    return is_valid
end
```

```julia
function validate_horizontal_symmetry(input_map, t)
    b = t+2
    t = t-1
    is_valid = true
    while (is_valid) && (t > 0) && (b < length(input_map)+1)
        is_valid = all(input_map[t] == input_map[b])
        t -= 1
        b += 1
    end
    return is_valid
end
```

```julia
function find_symmetry(current_map, old_val)
    output_val = 0
    # display(current_map)
    
    h_starts = find_potential_hline(current_map)

    for t in h_starts
        is_valid = validate_horizontal_symmetry(current_map, t)
        if (is_valid) && (old_val != -t)
            output_val = -t
            break
        end
    end
        
    v_starts = find_potential_vline(current_map)

    for l in v_starts
        is_valid = validate_vertical_symmetry(current_map, l)
        if (is_valid) && (old_val != l)
            output_val = l
            break
        end
    end

    return output_val
end
```
