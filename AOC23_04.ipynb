```julia
tot_sum = open("04_1.txt") do f
    nb_per_cards = Dict()
    wins_per_card = Dict()
    tot = 0
    idx = 1
    for l in eachline(f)
        l = replace(l, "  " => " ")
        nb = split(split(l, ": ")[2], " | ")
        winning_nbs = Set(split(nb[1], " "))
        drawn_nbs = Set(split(nb[2], " "))
        
        wins_per_card[idx] = length(intersect(winning_nbs, drawn_nbs))
        nb_per_cards[idx] = 1
        idx += 1
    end
    idx -= 1
    println(wins_per_card)
    for i in range(1, idx)
        for j in range(i+1, i+wins_per_card[i])
            nb_per_cards[j] += nb_per_cards[i]
        end
        tot += nb_per_cards[i]
    end
    println(nb_per_cards)
    (tot)
end
```
