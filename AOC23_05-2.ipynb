```julia
all_dicts_keys = ["seed-to-soil", "soil-to-fertilizer", "fertilizer-to-water", "water-to-light", "light-to-temperature", "temperature-to-humidity", "humidity-to-location"]
all_dicts = Dict()
for a_map_name in all_dicts_keys
    all_dicts[a_map_name] = Dict()
end
const seed_nb = Vector{Int}()
```

```julia
open("05_1.txt") do f
    min_l = 0
    # Read the seed numbers first 
    l = readline(f)
    seed_nb_vals = parse.(Int, split(l, ' ')[2:end])
    for a_val in seed_nb_vals
        push!(seed_nb, a_val)
    end
    readline(f)
    while !eof(f)
        l = readline(f)
        mapping_name = split(l, " map")[1]
        all_dicts[mapping_name]["destination_start"] = Vector{Int}()
        all_dicts[mapping_name]["source_start"] = Vector{Int}()
        all_dicts[mapping_name]["range"] = Vector{Int}()
            
        l = readline(f)
        while l != ""
            # Add a mapping
            values = split(l, ' ')
            push!(all_dicts[mapping_name]["destination_start"], parse(Int, values[1]))
            push!(all_dicts[mapping_name]["source_start"], parse(Int, values[2]))
            push!(all_dicts[mapping_name]["range"], parse(Int, values[3]))
            l = readline(f)
        end
    end
end
```

```julia
min_l = 1000000000
for i in seed_nb
    cur_val = i
    for a_mapping in all_dicts_keys
        new_val = cur_val
        # println("Current mapping name is $a_mapping with dict ", all_dicts[a_mapping])
        for (idx, a_map_range) in enumerate(all_dicts[a_mapping]["source_start"])
            if (cur_val >= a_map_range) && (cur_val < a_map_range + all_dicts[a_mapping]["range"][idx])
                new_val = all_dicts[a_mapping]["destination_start"][idx] + cur_val - a_map_range
            end
        end
        # println("Old value is $cur_val while new value is $new_val")
        cur_val = new_val
    end
    min_l = min(min_l, cur_val)
end
println(min_l)
```

```julia
seed_starts = seed_nb[1:2:end]
perm = sortperm(seed_starts)
seed_ranges = seed_nb[2:2:end]
```

```julia
function validate_seed(input_seed)
    idx = 1
    # println("Validating")
    # println(input_seed < seed_starts[perm[idx]])
    while (idx <= length(seed_starts)) && (input_seed >= seed_starts[perm[idx]])
        # println("Idx is $idx")
        # println("Input seed is $input_seed and seed start is ", seed_starts[perm[idx]])
        idx += 1
    end
    if idx == 1
        return false
    end
    idx -= 1
    # println("About to leave the function")
    # println(idx < length(seed_starts))
    # if idx <= length(seed_starts)
        # println((seed_starts[perm[idx]] <= input_seed))
        # println((input_seed < (seed_starts[perm[idx]] + seed_ranges[perm[idx]])))
    # end
    found_it = ((seed_starts[perm[idx]] <= input_seed) && (input_seed < (seed_starts[perm[idx]] + seed_ranges[perm[idx]])))
    return found_it
end
```

```julia
min_l = 0


found = false

while !found
    cur_val = min_l
    # println("TRYING VALUE $cur_val")
    for a_mapping_name in reverse(all_dicts_keys)
        new_val = cur_val
        # println("Current mapping name is $a_mapping_name with dict ", all_dicts[a_mapping_name])
        for (idx, a_map_range) in enumerate(all_dicts[a_mapping_name]["destination_start"])
            if (cur_val >= a_map_range) && (cur_val < a_map_range + all_dicts[a_mapping_name]["range"][idx])
                new_val = all_dicts[a_mapping_name]["source_start"][idx] + cur_val - a_map_range
                break
            end
        end
        # println("Old value is $cur_val while new value is $new_val")
        cur_val = new_val
    end
    # println("I am HERE!!!")
    found = validate_seed(cur_val)
    # println("HERE I am again!!!")
    
    if found 
        println("Current value is $min_l")
    end
    min_l += 1
    if (min_l % 5000) == 0
        println("min l is ", min_l)
    end
end
```
