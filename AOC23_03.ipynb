```julia
valid_symbols = "1234567890"
tot_sum = open("03_test.txt") do f
    tot = 0
    previous = "."^143
    current = "."^143
    for (idx_ln, next) in enumerate(eachline(f))
        next = '.'*next*'.'
        # println("Idx is $idx_ln and next line is $next while current is $current and previous is $previous")
        println("Idx is $idx_ln with line $current")
        for (idx_letter, c) in enumerate(current)
            if c == '*'
                nbs = Vector{Int}()
                if contains(valid_symbols, current[idx_letter-1])
                    end_idx = idx_letter - 1
                    start_idx = end_idx

                    while contains(valid_symbols, current[start_idx])
                        start_idx -= 1
                    end
                    start_idx += 1
                    push!(nbs, parse(Int, current[start_idx:end_idx]))
                end

                if contains(valid_symbols, current[idx_letter+1])
                    end_idx = idx_letter + 1
                    start_idx = end_idx

                    while contains(valid_symbols, current[end_idx])
                        end_idx += 1
                    end
                    end_idx -= 1
                    push!(nbs, parse(Int, current[start_idx:end_idx]))
                end

                for a_line in [previous, next]
                    do_center = true
                    do_right = true
                    if contains(valid_symbols, a_line[idx_letter-1])
                        do_center = false
                        end_idx = idx_letter - 1
                        start_idx = end_idx

                        while contains(valid_symbols, a_line[start_idx])
                            start_idx -= 1
                        end
                        start_idx += 1

                        if contains(valid_symbols, a_line[idx_letter])
                            do_right = false
                            end_idx = idx_letter
                            while contains(valid_symbols, a_line[end_idx])
                                end_idx += 1
                            end
                            end_idx -= 1
                        end
                        push!(nbs, parse(Int, a_line[start_idx:end_idx]))
                    end

                    if do_center && contains(valid_symbols, a_line[idx_letter])
                        # We can't go left
                        start_idx = idx_letter 
                        end_idx = start_idx
                        do_right = false
                        while contains(valid_symbols, a_line[end_idx])
                            end_idx += 1
                        end
                        end_idx -= 1
                        push!(nbs, parse(Int, a_line[start_idx:end_idx]))
                    end

                    if do_right && contains(valid_symbols, a_line[idx_letter+1])
                        start_idx = idx_letter + 1 
                        end_idx = start_idx
                        do_right = false
                        while contains(valid_symbols, a_line[end_idx])
                            end_idx += 1
                        end
                        end_idx -= 1
                        push!(nbs, parse(Int, a_line[start_idx:end_idx]))
                    end
                        
                    
                        
                end
                if length(nbs) == 2
                    tot += nbs[1]*nbs[2]
                end
            end # Of 'c' character 
        end # of all letters
        previous = current 
        current = next
    end # of al lines
        
    
    (tot)
end
```
