```julia
function load_line(line)
	map(collect(line)) do c
		if c == '#'
			2 # 'Loaded' rock
		elseif c == 'O'
			1 'Rolling' rock
		else
			0
		end
	end
end
```

```julia
function go_north(dish)
    for j in range(1, length(dish[1]))
        last_loaded_rock = 0
        for i in range(1, length(dish))
            if dish[i][j] == 2
                last_loaded_rock = i
            elseif dish[i][j] == 1
                dish[i][j] = 0
                last_loaded_rock += 1
                dish[last_loaded_rock][j] = 1
            end
        end
    end

    return dish
        
end
```

```julia
(tot_load) = open("14_1.txt") do f 
    tot = 0
    mirror_map = []
    for l in eachline(f)
        push!(mirror_map, load_line(l))
    end
    nb_rows = length(mirror_map)
    
    N_map = go_north(mirror_map)

    for (idx, a_row) in enumerate(N_map)
        #println(a_row)
        #println(findall(a_row .== 1))
        tot += (nb_rows+1-idx)*sum(a_row .== 1)
    end
    (tot)
end
```

```julia
function go_north_or_not(dish, go_north=true)
    for j in range(1, length(dish[1]))
        last_loaded_rock = go_north ? 0 : length(dish)+1
        if go_north
            ranges = range(1, length(dish))
        else
            ranges = length(dish):-1:1
        end
        for i in ranges
            if dish[i][j] == 2
                last_loaded_rock = i
            elseif dish[i][j] == 1
                dish[i][j] = 0
                last_loaded_rock = go_north ? last_loaded_rock + 1 : last_loaded_rock - 1
                dish[last_loaded_rock][j] = 1
            end
        end
    end

    return dish
        
end
```

```julia
function go_west_or_not(dish, go_west=true)
    for i in range(1, length(dish))
        last_loaded_rock = go_west ? 0 : length(dish[1])+1
        if go_west
            ranges = range(1, length(dish[1]))
        else
            ranges = length(dish[1]):-1:1
        end
        for j in ranges
            if dish[i][j] == 2
                last_loaded_rock = j
            elseif dish[i][j] == 1
                dish[i][j] = 0
                last_loaded_rock = go_west ? last_loaded_rock + 1 : last_loaded_rock - 1
                dish[i][last_loaded_rock] = 1
            end
        end
    end

    return dish
        
end
```

```julia
function tilt_four_ways(dish)
    dish = go_north_or_not(dish)
    dish = go_west_or_not(dish)
    dish = go_north_or_not(dish, false)
    dish = go_west_or_not(dish, false)
    return dish 
end
```

```julia
function la_fontaine_cycle(dish, f_handler)
    t = deepcopy(dish) # The tortoise
    h = deepcopy(dish) # The haze
    t = f_handler(t)
    h = f_handler(f_handler(h))
    nb_t = 1
    while t != h
        t = f_handler(t)
        h = f_handler(f_handler(h))
        nb_t += 1
    end
    # nb_t is divisible by the period of the cycle. 
    # Let's figure out when the cycle start 
    
    start_point = 0
    t = deepcopy(dish)
    while t != h
        t = f_handler(t)
        h = f_handler(h)
        start_point += 1
    end

    # Now, find the shortest cycle
    c = 1
    h = f_handler(h)
    while t != h
        c += 1
        h = f_handler(h)
    end
    return (c, start_point)
end
```
