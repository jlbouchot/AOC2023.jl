```julia
(prod_el) = open("25_1.txt") do f

    graph = DefaultDict(() -> [])

    for l in eachline(f) 
        values = split(l, ": ")
        lhs = values[1]
        for rhs in split(values[2], " ")
            push!(graph[lhs], rhs)
            push!(graph[rhs], lhs)
        end
    end
    println(graph)

    
    names_to_idx = Dict()
    idx_to_names = Dict()

    for (idx, name) in enumerate(keys(graph))
        names_to_idx[name] = idx
        idx_to_names[idx] = name
    end

    # Create the adjacency matrix, hoping to have enough RAM
    n = length(keys(graph))
    adj_mat = zeros(Int32, n, n)
    println(names_to_idx)

    for (idx, name) in enumerate(keys(graph))
        for a_connection in graph[name]
            adj_mat[names_to_idx[name], names_to_idx[a_connection]] = 1
        end
    end

    display(adj_mat)

    lap = Matrix(1I, n, n) 
    for (idx,i) in enumerate(sum(adj_mat, dims = 1))
        lap[idx,idx] = i
    end
    lap = lap - adj_mat

    display(lap)

    evals = eigvals(lap)
    evect = eigvecs(lap)

    println(evals)
    fiedler = evect[:,2]
    println(fiedler)

    npos = length(findall(x -> x > 0, fiedler))
    out = npos*(n-npos)

    (out)

end
```
