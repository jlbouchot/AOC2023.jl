```julia
function which_way(directions)
    dir_mapping = Dict(('R', 'D') => -1, ('R', 'U') => +1, ('L', 'U') => -1, ('L', 'D') => +1, ('U', 'R') => -1, ('U', 'L') => +1, ('D', 'L') => -1, ('D', 'R') => +1)
    loc_dir = counter([dir_mapping[(directions[i-1], directions[i])] for i in range(2, length(directions))])
    largest_dir = -1
    if loc_dir[1] > loc_dir[-1]
        largest_dir = 1
    end
    return largest_dir
end
```

```julia
(nb_m3) = open("18_1.txt") do f 
    nb_m3 = 0
    
    directions = []
    steps = []
    colours = []
    for l in eachline(f)
        values = split(l, ' ')
        push!(directions, only(values[1]))
        push!(steps, parse(Int, values[2]))
        push!(colours, values[3])
    end

    main_dir = which_way(directions)

    grid_size = sum(steps)
    nb_m3 += grid_size
    grid = fill(false, (2*grid_size, 2*grid_size))

    cur_row = grid_size
    cur_col = grid_size

    need_a_visit = []

    for i in range(1, length(steps))
        # println("Current position is $cur_row,$cur_col. Looking at directions $(directions[i]) and steps size $(steps[i])")
        if directions[i] == 'R'
            grid[cur_row,cur_col+1:cur_col+steps[i]] .= true
            delta = main_dir > 0 ? -1 : +1
            for i in range(1, steps[i])
                push!(need_a_visit, [cur_row+delta, cur_col+i])
            end
            cur_col += steps[i]
        elseif directions[i] == 'L'
            grid[cur_row,cur_col-steps[i]:cur_col-1] .= true
            delta = main_dir > 0 ? +1 : -1
            for i in range(1, steps[i])
                push!(need_a_visit, [cur_row+delta, cur_col-i])
            end
            cur_col -= steps[i]
        elseif directions[i] == 'D'
            grid[cur_row+1:cur_row+steps[i],cur_col] .= true
            delta = main_dir > 0 ? +1 : -1
            for i in range(1, steps[i])
                push!(need_a_visit, [cur_row+i, cur_col+delta])
            end
            cur_row += steps[i]
        else
            grid[cur_row-steps[i]:cur_row-1,cur_col] .= true
            delta = main_dir > 0 ? -1 : +1
            for i in range(1, steps[i])
                push!(need_a_visit, [cur_row-i, cur_col+delta])
            end
            cur_row -= steps[i]
        end
        # println(need_a_visit)
    end
    # println(need_a_visit)
    # display(grid[30:50,30:50])

    while length(need_a_visit) > 0
        current_cell = pop!(need_a_visit)
        # println("Current cell is $current_cell , and grid location gives $(grid[current_cell])")
        if !(grid[current_cell[1],current_cell[2]])
            nb_m3 += 1
            push!(need_a_visit, current_cell+[0,1])
            push!(need_a_visit, current_cell+[0,-1])
            push!(need_a_visit, current_cell+[1,0])
            push!(need_a_visit, current_cell+[-1,0])
            grid[current_cell[1],current_cell[2]] = true
            # display(grid[30:50,30:50])
        end
    end

    (nb_m3)
end
```

```julia
function parse_input(current_colour)
    directon_mapping = Dict('0' => 'R', '1' => 'D', '2' => 'L', '3' => 'U')
    return (parse(Int,current_colour[2:end-1], base=16), directon_mapping[current_colour[end]])
end
```
