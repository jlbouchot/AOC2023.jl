```julia
function find_first_steps(idx_row, idx_col, pipe_map)
    sols = []
    size_in = size(pipe_map)
    if (idx_col > 1) && (pipe_map[idx_row,idx_col-1] in ['F', '-', 'L'])
         push!(sols, [idx_row,idx_col-1])
    end
    if (idx_col < size_in[2]) && (pipe_map[idx_row,idx_col+1] in ['J', '-', '7'])
        push!(sols, [idx_row,idx_col+1])
    end 
    if (idx_row > 1) && (pipe_map[idx_row-1,idx_col] in ['|', '7', 'F'])
       push!(sols, [idx_row-1,idx_col])
    end 
    if (idx_row < size_in[1]) && (pipe_map[idx_row+1,idx_col] in ['|', 'J', 'L'])
       push!(sols, [idx_row+1,idx_col])
    end 
    return sols
end
```

```julia
function find_next_steps(idx_row, idx_col, current_char)
    out = []
    if current_char == '-'
        out = [[idx_row, idx_col-1], [idx_row, idx_col+1]]
    elseif current_char == '|'
        out = [[idx_row-1,idx_col], [idx_row+1, idx_col]]
    elseif current_char == 'J'
        out = [[idx_row,idx_col-1], [idx_row-1, idx_col]]
    elseif current_char == 'L' 
        out = [[idx_row, idx_col+1], [idx_row-1, idx_col]]
    elseif current_char == 'F'
        out = [[idx_row, idx_col+1], [idx_row+1, idx_col]]
    elseif current_char == '7'
        out = [[idx_row, idx_col-1], [idx_row+1, idx_col]]
    end
    return out
end
```

```julia
function which_way(past, current, future)
    out = 0
    if future[1] - past[1] == 1
        if current[1] == future[1] 
            out = 1 # Turn left
        else
            out = 2 # Turn right
        end
    end
    return out
end
```

```julia
nb_inside = open("10_1.txt") do f 
    l = readlines(f)
    nb_rows = length(l)
    nb_cols = length(l[1])
    pipe_map = Array{Char}(undef, nb_rows, nb_cols)
    for idx_row in range(1, nb_rows)
        pipe_map[idx_row, :] = collect(l[idx_row])
    end

    idx_S = findall(x -> x == 'S', pipe_map)[1]

    dist_map = Array{Int}(undef, nb_rows, nb_cols)
    visited = [[idx_S[1], idx_S[2]]]
    for idx_row in range(1, nb_rows)
        for idx_col in range(1, nb_cols)
            dist_map[idx_row, idx_col] = 0
        end
    end

    potential_directions = find_first_steps(idx_S[1], idx_S[2], pipe_map)
    past_loc = [idx_S[1], idx_S[2]]
    current_loc = potential_directions[1]
    push!(visited, current_loc)
    dist_map[current_loc[1], current_loc[2]] = 1
    turned_left = 0
    turned_right = 0

    ways = []

    while (current_loc != visited[1])
        potential_directions = find_next_steps(current_loc[1], current_loc[2], pipe_map[current_loc[1], current_loc[2]])
        if length(potential_directions) == 1
            push!(potential_directions, visited[1])
        end
        if potential_directions[1] == past_loc
            way = which_way(past_loc, current_loc, potential_directions[2])
            past_loc = current_loc 
            current_loc = potential_directions[2]
        else
            way = which_way(past_loc, current_loc, potential_directions[1])
            past_loc = current_loc
            current_loc = potential_directions[1]
        end
        push!(visited, current_loc)
        dist_map[current_loc[1], current_loc[2]] = dist_map[past_loc[1], past_loc[2]] + 1
        if way == 1
            turned_left += 1
        else
            turned_right += 1
        end
        push!(ways, way)
    end

    display(dist_map)

    went_right = (turned_right > turned_left)
    println(went_right)

    dequeue = []
    past = visited[1]
    for current in visited[2:end]
        delta = current - past
        to_the_inside = [0, 0]
        if delta[1] > 0
            # println("Going downwards")
            to_the_inside[2] = - 1
        elseif delta[1] < 0 
            # println("Going upwards") 
            to_the_inside[2] = + 1
        elseif delta[2] > 0
            # println("Gong to the right")
            to_the_inside[1] = + 1
        else
            # println("Going to the left")
            to_the_inside[1] = - 1
        end
        println("Past is $past current is $current delta is $delta and were adding $to_the_inside")
        if went_right
            push!(dequeue, current + to_the_inside)
            push!(dequeue, past + to_the_inside)
        else
            push!(dequeue, current - to_the_inside)
            push!(dequeue, past - to_the_inside)
        end
        past = current
        #println(dequeue)
    end

    # println("Visited is $visited")

    is_inside = []
    while length(dequeue) > 0 
        # println(dequeue)
        current = pop!(dequeue)
        # println("$current is visited? $(current in visited)")
        if (current in visited) || (current in is_inside) || (current[1] == 0) || (current[2] == 0) || (current[1] == nb_rows+1) || (current[2] == nb_cols+1)
            continue
        end
        push!(is_inside, current)
        if current[1] > 1
            push!(dequeue, current + [-1,0])
        end
        if current[2] > 1
            push!(dequeue, current + [0,-1])
        end
        if current[1] < nb_rows
            push!(dequeue, current + [1,0])
        end
        if current[2] < nb_cols
            push!(dequeue, current + [0,1])
        end

        # println("Is inside is ", is_inside)
            
    end
    

    nb_inside = length(is_inside)
    
    (nb_inside)

end
```
