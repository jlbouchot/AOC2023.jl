```julia
function find_string_numerals(s_in) 
    delta = 1
    lead = s_in[1:1]
    if startswith(s_in, "one")
        lead = "1"
        delta = 3
    end
    if startswith(s_in, "two")
        lead = "2"
        delta = 3
    end
    if startswith(s_in, "three")
        lead = "3"
        delta = 5
    end
    if startswith(s_in, "four")
        lead = "4"
        delta = 4
    end
    if startswith(s_in, "five")
        lead = "5"
        delta = 4
    end
    if startswith(s_in, "six")
        lead = "6"
        delta = 3
    end
    if startswith(s_in, "seven")
        lead = "7"
        delta = 5
    end
    if startswith(s_in, "eight")
        lead = "8"
        delta = 5
    end
    if startswith(s_in, "nine")
        lead = "9"
        delta = 4
    end
    if length(s_in) <= delta 
        return lead
    else
        return lead * find_string_numerals(s_in[delta+1:end])
    end
end
    
        
```

```julia
function find_first_digit(s_in)
    dict_nb = Dict("one" => '1', "two" => '2', "three" => '3', "four" => '4', "five" => '5', "six" => '6', "seven" => '7', "eight" => '8', "nine" => '9')
    v = ""
    if startswith(s_in, ['1','2','3','4','5','6','7','8','9'])
        v = s_in[1:1]
    elseif startswith(s_in, "one") || startswith(s_in,"two") || startswith(s_in, "six")
        v = dict_nb[s_in[1:3]]
    elseif startswith(s_in, "four") || startswith(s_in,"five") || startswith(s_in, "nine")
        v = dict_nb[s_in[1:4]]
    elseif startswith(s_in, "three") || startswith(s_in,"seven") || startswith(s_in, "eight")
        v = dict_nb[s_in[1:5]]
    else
        return find_first_digit(s_in[2:end])
    end
    
    return parse(Int64, v)
end
```

```julia
function find_last_digit(s_in)
    dict_nb = Dict("one" => '1', "two" => '2', "three" => '3', "four" => '4', "five" => '5', "six" => '6', "seven" => '7', "eight" => '8', "nine" => '9')
    if endswith(s_in, ['1','2','3','4','5','6','7','8','9'])
        v = s_in[end:end]
    elseif endswith(s_in, "one") || endswith(s_in,"two") || endswith(s_in, "six")
        v = dict_nb[s_in[end-2:end]]
    elseif endswith(s_in, "four") || endswith(s_in,"five") || endswith(s_in, "nine")
        v = dict_nb[s_in[end-3:end]]
    elseif endswith(s_in, "three") || endswith(s_in,"seven") || endswith(s_in, "eight")
        v = dict_nb[s_in[end-4:end]]
    else
        return find_last_digit(s_in[1:end-1])
    end
    
    return parse(Int64, v)
end
```
